```python
import pandas as pd
import os
import numpy as np
```

# Carrega a base de dados

```python
df = pd.read_csv('kick1.dat', sep = ' ', header = None, names = ['x', 'y', 'z'])
df
```

# Calcula as derivadas de cada coordenada

```python
df['x_diff'] = df['x'].diff().fillna(0)
df['y_diff'] = df['y'].diff().fillna(0)
df['z_diff'] = df['z'].diff().fillna(0)
df
```

```python
LaggingSamples = 2
```

```python
InputColumnNames = []
for i in range(1, LaggingSamples+1):
    InputColumnNames.append('x-' + str(i))
    InputColumnNames.append('y-' + str(i))
    InputColumnNames.append('z-' + str(i))
    InputColumnNames.append('x_diff-' + str(i))
    InputColumnNames.append('y_diff-' + str(i))
    InputColumnNames.append('z_diff-' + str(i))
```

```python
Input = []
Output = []
for i in range(LaggingSamples, len(df)):
    AuxInput = []
    for j in range(LaggingSamples):
        AuxInput.append(df.loc[i - (j + 1), :].tolist())
    AuxInput = sum(AuxInput, [])
    Input.append(AuxInput)
    Output.append(df.loc[i, :].tolist())
```

```python
Input = pd.DataFrame(Input, columns = InputColumnNames)
Output = pd.DataFrame(Output, columns = ['x', 'y', 'z', 'x_diff', 'y_diff', 'z_diff'])
```

```python
Input
```

```python
from sklearn.tree import DecisionTreeRegressor

model = DecisionTreeRegressor()

model.fit(Input, Output)
```

# Inicia a previsão da trajetória

```python
Data = Input.loc[0, :].tolist()

Outputs = []
for i in range(len(df) - LaggingSamples):
    Out = model.predict(np.array(Data).reshape(1, -1))
    Data = Data[6:]
    for j in range(len(Out[0])):
        Data.append(Out[0][j])
    
    Outputs.append(Out[0])

```

```python
Outputs = pd.DataFrame(Outputs, columns = ['x', 'y', 'z', 'x_diff', 'y_diff', 'z_diff'])
```
