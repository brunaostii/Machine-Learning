```python
import pandas as pd
from sklearn import linear_model
from sklearn.model_selection import train_test_split


kick1 = pd.read_csv('kick1.dat', sep = ' ', header = None, names = ['x', 'y', 'z'])
kick2 = pd.read_csv('kick2.dat', sep = ' ', header = None, names = ['x', 'y', 'z'])
```

```python
import matplotlib.pyplot as plt 
import numpy as np

plt.figure(figsize=(8,5))
ax = plt.axes(projection='3d')
ax.plot([-3, -3, 3, 3], [0, 0.3, 0.3, 0], zdir='y', zs=0, c='b')
X, Y = np.meshgrid(np.linspace(-3, 3, 2), np.linspace(0,2,2))
Z = np.zeros(X.shape)
ax.plot_surface(X, Y, Z, shade = False, color='g', alpha = .4)
ax.scatter3D(kick1['x'], kick1['y'], kick1['z'])
plt.show()
```

```python
plt.figure(figsize=(8,5))
ax = plt.axes(projection='3d')
ax.plot([-3, -3, 3, 3], [0, 0.3, 0.3, 0], zdir='y', zs=0, c='b')
X, Y = np.meshgrid(np.linspace(-3, 3, 2), np.linspace(0,2,2))
Z = np.zeros(X.shape)
ax.plot_surface(X, Y, Z, shade = False, color='g', alpha = .4)
ax.scatter3D(kick2['x'], kick2['y'], kick2['z'])
plt.show()
```

```python
time = np.linspace(0, 1, 60)
```

```python
time
```

### 'X'

```python
X_train, X_test, y_train, y_test = train_test_split(time[:20].reshape(-1, 1), kick1['x'], test_size=0.2, random_state=42)

reg = linear_model.LinearRegression()

reg.fit(X_train, y_train)
```

```python
reg.predict(X_test)
```

```python
y_test
```

```python
x_pred = reg.predict(time[:43].reshape(-1, 1))
```

### 'Y'

```python
X_train, X_test, y_train, y_test = train_test_split(time[:20].reshape(-1, 1), kick1['y'], test_size=0.2, random_state=42)

reg = linear_model.LinearRegression()

reg.fit(X_train, y_train)
```

```python
reg.predict(X_test)
```

### 'Z'

```python
X_train, X_test, y_train, y_test = train_test_split(time[:20].reshape(-1, 1), kick1['z'], test_size=0.2, random_state=42)

reg = linear_model.LinearRegression()

reg.fit(X_train, y_train)
```

```python
reg.predict(X_test)
```

```python
y_test
```

```python
z_pred = reg.predict(time[:43].reshape(-1, 1))
```

```python
plt.figure(figsize=(20,10))
ax = plt.axes(projection='3d')
ax.plot([-3, -3, 3, 3], [0, 0.3, 0.3, 0], zdir='y', zs=0, c='b')
X, Y = np.meshgrid(np.linspace(-3, 3, 2), np.linspace(0,2,2))
Z = np.zeros(X.shape)
ax.plot_surface(X, Y, Z, shade = False, color='g', alpha = .4)
ax.scatter3D(kick1['x'], kick1['y'], kick1['z'])
ax.plot(x_pred, y_pred, z_pred)
plt.show()
```
