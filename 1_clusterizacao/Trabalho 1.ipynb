## Imagens coloridas

```python
import numpy as np
import cv2 #scipy with imread has been deprecated 
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('./images/rgb/baboon.png')
```

```python
type(img)
```

```python
img.shape 
```

```python
red = img[:, :, 2]
green = img[:, :, 1]
blue = img[:, :, 0]
```

```python
plt.imshow(red, cmap='Reds')
```

```python
plt.imshow(green, cmap='Greens')
```

```python
plt.imshow(blue, cmap='Blues')
```

```python
r = 0.393*red + 0.769*green + 0.189*blue
g = 0.349*red + 0.686*green+ 0.168*blue
b = 0.272*red + 0.534*green+ 0.131*blue

img_rec = np.zeros(img.shape, dtype='uint8')
```

```python
img_rec[:,:,0] = r
img_rec[:,:,1] = g
img_rec[:,:,2] = b
```

```python
img_rec[img_rec > 255] = 255
```

```python
plt.imshow(img_rec)
```
